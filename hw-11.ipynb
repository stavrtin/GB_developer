```python
# f(x) = x^2 - 6*abs(x) + 8
```

```python
from sympy import *
from sympy.plotting import plot
init_printing()
```

```python
x = Symbol('x', real=True)
f = x**2 - 6 * abs(x) + 8
f
```

```python
plot(f)
```

```python
# корни уравнения
root = solve(f)
root
```

```python
#  ищем производную
differ = diff(f)
differ
```

```python
res_dif = solve(differ)
res_dif
```

```python
plot(differ)
```

```python
res_dif = solve(differ)
lst = [-3, 0, *res_dif]
lst
```

```python
solve(differ > 0)
```

```python
lst = []
lst.append((res_dif, 3,0))
    
lst  
    
```

```python
# где функция больше нуля
solve(f > 0)
```

```python
# где функция меньше нуля
solve(а < 0)
```

```python
# положительна на этом
solve(differ > 0)
```

```python
# отрицательна на этом
solve(differ < 0)
```

```python
# вершины

extrem = []
for i in lst:
    extrem.append((i, f.subs(x,i)))
extrem
    
```
